```python
# -*- coding:utf-8 -*-
import re
import requests
import urllib

i = 1
urlList = []

def dowmloadPic(html, keyword):
    global i
    pic_url = re.findall('"objURL":"(.*?)",', html, re.S)
    #print(pic_url)
    print('找到关键词:' + keyword + '的图片，现在开始下载图片...')
    for each in pic_url:
        if str(each) in urlList:
            continue
            
        print('正在下载第' + str(i) + '张图片，图片地址:' + str(each))
        try:
            pic = requests.get(each, timeout=5)
        except requests.exceptions.ConnectionError:
            print('【错误】当前图片无法下载')
            continue
        except requests.exceptions.ReadTimeout:
            print('【错误】连接超时')
            continue
        #except Exception:
           # print('【错误】其它错误')
            #continue

        dir = 'lotus/' + keyword + '_' + str(i) + '.jpg'
        fp = open(dir, 'wb')
        fp.write(pic.content)
        
        dir = 'lotus/' + keyword + '_' + str(i) + '.json'
        fp = open(dir, 'w')
        fp.write("{\"labels\": [{\"name\": \"lotus\"}]}")
        
        fp.close()
        i += 1
        urlList.append(str(each))
    
if __name__ == '__main__':
    word = input("Input key word: ")
    #url = 'http://image.baidu.com/search/index?tn=baiduimage&ps=1&ct=201326592&lm=-1&cl=2&nc=1&ie=utf-8&word=' + word 
    #url = 'http://image.baidu.com/search/flip?tn=baiduimage&ie=utf-8&word=' + word + '&ct=201326592&v=flip'
    url='http://image.baidu.com/search/flip?tn=baiduimage&ie=utf-8&word={0}&pn={1}'
    page_num = 20
    num = 0
    #url = 'https://image.baidu.com/search/acjson?tn=resultjson_com&ipn=rj&ct=201326592&is=&fp=result&queryWord=&cl=2&lm=-1&ie=utf-8&oe=utf-8&adpicid=&st=-1&z=&ic=&hd=&latest=&copyright=&word={0}&s=&se=&tab=&width=&height=&face=0&istype=2&qc=&nc=1&fr=&expermode=&force=&pn={1}&rn={2}&gsm=1e&1552975216767='
   
    while num < 100:
        print("循环测试：" + str(num))
        page_url=url.format(urllib.parse.quote(word),num*page_num)
        result = requests.get(page_url)
        print(page_url)
        dowmloadPic(result.text, word)
        num += 1
```
